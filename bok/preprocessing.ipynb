```python
value_counts = df['subCategory'].value_counts()
total_count = len(df['subCategory'])

# 각 값의 갯수와 비율 계산
value_ratio = pd.DataFrame({'Count': value_counts, 'Ratio': value_counts / total_count})
# value_ratio.to_excel('value_ratio.xlsx')
value_ratio
```
