```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
# 파일경로
img_path = "./garbage_classification/"
```

### 상하반전

```python
# img_path 내의 모든 폴더에 대해 반복
for root, dirs, files in os.walk(img_path):
    # 폴더 내의 각 파일에 대해 반복하여 경로 출력
    for file in files:
        file_path = os.path.join(root, file)
        img = cv2.imread(file_path)
        # print(file_path)
        # 좌우 반전
        img_flip_ud = cv2.flip(img, 1)

        # 변환된 파일 보기
        # plt.imshow(img_flip_ud)
        # break

        new_file_path = file_path.replace('tion','tion2').replace('.jpg','ud.jpg')
        # print(new_file_path)
        cv2.imwrite(new_file_path, img_flip_ud)
```

### 좌우반전

```python
# img_path 내의 모든 폴더에 대해 반복
for root, dirs, files in os.walk(img_path):
    # 폴더 내의 각 파일에 대해 반복하여 경로 출력
    for file in files:
        file_path = os.path.join(root, file)
        img = cv2.imread(file_path)
        # print(file_path)
        # 좌우 반전
        img_flip_lr = cv2.flip(img, 0)

        # 변환된 파일 보기
        # plt.imshow(img_flip_lr)
        # break

        new_file_path = file_path.replace('tion','tion2').replace('.jpg','lr.jpg')
        # print(new_file_path)
        cv2.imwrite(new_file_path, img_flip_lr)
```

### 상하좌우 반전

```python
# img_path 내의 모든 폴더에 대해 반복
for root, dirs, files in os.walk(img_path):
    # 폴더 내의 각 파일에 대해 반복하여 경로 출력
    for file in files:
        file_path = os.path.join(root, file)
        img = cv2.imread(file_path)
        
        # 상하좌우 반전
        img_flip_both = cv2.flip(img, -1)

        # 변환된 파일 보기
        # plt.imshow(img_flip_both)
        # break

        new_file_path = file_path.replace('tion','tion2').replace('.jpg','udlr222222222.jpg')
        print(new_file_path)
        cv2.imwrite(new_file_path, img_flip_both)
```

### 상하 반전 + 흑백 변환

```python
# img_path 내의 모든 폴더에 대해 반복
for root, dirs, files in os.walk(img_path):
    # 폴더 내의 각 파일에 대해 반복하여 경로 출력
    for file in files:
        file_path = os.path.join(root, file)
        img = cv2.imread(file_path)
        
        # 상하 반전
        img_flip_ud = cv2.flip(img, 0)
        # 흑백 변환
        gray_img = cv2.cvtColor(img_flip_ud, cv2.COLOR_RGB2GRAY)
        
        # 변환된 파일 보기
        # plt.imshow(gray_img, cmap='grey')
        # break

        new_file_path = file_path.replace('tion','tion2').replace('.jpg','gray.jpg')
        # print(new_file_path)
        cv2.imwrite(new_file_path, gray_img)
```
